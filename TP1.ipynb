```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')

df = pd.read_csv("customer_support_twitter.csv")
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    text_without_stopwords = ' '.join(filtered_words)
    return text_without_stopwords
df['text'] = df['text'].apply(remove_stopwords)
print(df.head())
```

```python
import pandas as pd
from collections import Counter

df = pd.read_csv("customer_support_twitter.csv")
all_text = ' '.join(df['text'])
words = all_text.split()
word_freq = Counter(words)

num_freq_words_to_remove = 10
most_common_words = word_freq.most_common(num_freq_words_to_remove)

most_common_words = [word for word, freq in most_common_words]

def remove_common_words(text):
    words = text.split()
    filtered_words = [word for word in words if word not in most_common_words]
    text_without_common_words = ' '.join(filtered_words)
    return text_without_common_words
df['text'] = df['text'].apply(remove_common_words)
print(df.head())
```

```python
import pandas as pd
from collections import Counter

df = pd.read_csv("customer_support_twitter.csv")

all_text = ' '.join(df['text'])

words = all_text.split()

word_freq = Counter(words)

min_word_freq = 2

rare_words = [word for word, freq in word_freq.items() if freq < min_word_freq]
def remove_rare_words(text):
    words = text.split()
    filtered_words = [word for word in words if word not in rare_words]
    text_without_rare_words = ' '.join(filtered_words)
    return text_without_rare_words

df['text'] = df['text'].apply(remove_rare_words)

print(df.head())
```

```python
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('wordnet')
df = pd.read_csv("customer_support_twitter.csv")

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    lemmatized_text = ' '.join(lemmatized_tokens)
    return lemmatized_text

df['text'] = df['text'].apply(lemmatize_text)

print(df.head())
```
